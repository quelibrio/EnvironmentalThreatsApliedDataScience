```python
import csv
exampleFile = open('W-2016-M.csv')
#exampleFile = open('2016-M.csv')
exampleReader = csv.reader(exampleFile)
exampleData = list(exampleReader)
```

```python
#exampleData[0][11]
#exampleData[0][12]
maxRows = len(exampleData)
exampleData[0][1]
```

```python
#newData = []
X = []
Y = []
Z = []
minValue=999999999.0
maxValue=0.0
for i in range(1,len(exampleData)):
#for i in range(1,50):
    #new = []
    #new.append(exampleData[i][11])
    #new.append(exampleData[i][12])
    #newData.append(new)
    if exampleData[i][1] == 'Oxygen Diss':
        X.append(exampleData[i][11])
        Y.append(exampleData[i][12])
        Z.append(float(exampleData[i][5]))
        if float(exampleData[i][5])>maxValue:
            maxValue=float(exampleData[i][5])
        if float(exampleData[i][5])<minValue:
            minValue=float(exampleData[i][5])
        #colourmap.append('red')
fluct=maxValue-minValue
print ('maxValue')
print (maxValue)
print ('minValue')
print (minValue)
#newData
#print ('X')
#print (X)
#===================
colourmap = []
colorType=1 #0: purple gradient, 1: green, blue gradient
for i in range(0,len(Z)):
    dist=(Z[i]-minValue)/fluct
    g=0
    r=dist
    b=1.0-dist
    if colorType==1:
        if dist<0.5:
            dist=dist*2
            g=dist*0.5
            r=dist*0.5
            b=1-dist*0.5
        else:
            dist=1.0-dist
            dist=dist*2
            g=dist*0.5
            b=dist*0.5
            r=1-dist*0.5
    colourmap.append((r, g, b))
```

```python
from matplotlib import pyplot as plt
plt.scatter(X,Y,c=colourmap, lw = 0)
plt.show()
```
